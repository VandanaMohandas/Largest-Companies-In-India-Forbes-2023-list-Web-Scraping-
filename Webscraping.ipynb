```python


# BASIC WEBSCARPING


from bs4 import BeautifulSoup  # librray to scrape text
import requests                 # library to get the website content
import csv
page_to_scrape = requests.get("https://quotes.toscrape.com") #Loading it into a variable the website
soup =BeautifulSoup(page_to_scrape.text,"html.parser")       # scraping

#Next step ,go to the website inspect the elemet and understand the calss of the element one has to scrape
Quotes =soup.findAll("span",attrs={"class":"text"})
Author =soup.findAll("small",attrs={"class":"author"})


file = open("scraped_file.csv","w")
writer=csv.writer(file)
writer.writerow(["Quotes","Authors"])


for q,a in zip(Quotes,Author):
  print(q.text+ "--" + a.text)
  writer.writerow([q.text,a.text])
file.close()
```

**Wikipedia- Webscrape  **-Largest Comapnies In India Forbes 2023 list

```python
from bs4 import BeautifulSoup
import requests
import csv
import pandas as pd
website=requests.get("https://en.wikipedia.org/wiki/List_of_largest_companies_in_India")
soup=BeautifulSoup(website.text,"html.parser")
soup.find_all('table')[0]
table = soup.find_all('table',class_='wikitable sortable')[0]
table_header = table.find_all('th')
table_header1 = [t.text.strip() for t in table_header]
table_header1
df=pd.DataFrame(columns = table_header1)
Main_data= table.find_all('tr')
for m in Main_data[1::]:
  rd = m.find_all('td')
  row_data =[d.text.strip() for d in rd]
  length=len(df)
  df.loc[length] =row_data
df




```
